```python
################################################################
# pip install pandas-profiling
from pandas_profiling import ProfileReport
import pandas as pd

df = pd.read_csv('amazon_reviews_us_Shoes_v1_00_help_voted_And_cut_lognTail.csv')


profile = ProfileReport(df, title='amazon_reviews_us_Shoes_v1_00_help_voted_And_cut_lognTail', minimal=True)
profile.to_file(output_file="amazon_reviews_us_Shoes_v1_00_help_voted_And_cut_lognTail.html")
profile.to_notebook_iframe()
################################################################
```
