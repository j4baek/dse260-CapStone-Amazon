```python
################################################################
# pip install pandas-profiling
from pandas_profiling import ProfileReport
import pandas as pd

df = pd.read_csv('Shoes_for_100_users_per_10_products_prediction.csv')


profile = ProfileReport(df.drop(columns=['user_id']), title='Shoes_for_100_users_per_10_products_prediction-Report', minimal=True)
profile.to_file(output_file="Shoes_for_100_users_per_10_products_prediction-Report.html")
profile.to_notebook_iframe()
################################################################
```
